```python
df = pd.read_excel("/Users/elmoore/Documents/Python Data Wrangling Class/Sample - Superstore.xls")
df.head()
```

# pivot table

```python
df_sample = df.sample(n=100)
df_sample
```

```python
pt = df_sample.pivot_table(
    values=['Sales'], 
    columns='Category',
    index=['Region'],
    aggfunc='sum')
pt
```

```python
pt.plot(use_index=True, y=pt.columns, kind='bar')
```
