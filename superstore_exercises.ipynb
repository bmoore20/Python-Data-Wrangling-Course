# Exercise 59: Flexibility for User-Defined Functions with the apply Method

```python
def categorize_sales(price): 
    if price < 50:
        return "Low" 
    elif price < 200:
        return "Medium" 
    else:
        return "High"
```

```python
df_sample=df[['Customer Name','State','Sales']].sample(n=100)
df_sample.head(10)
```

```python
df_sample['Sales Price Category']=df_sample['Sales'].apply(categorize_sales)
df_sample.head(10)
```

```python
df_sample['Customer Name Length']=df_sample['Customer Name'].apply(len)
df_sample.head(10)
```

```python
df_sample['Discounted Price']=df_sample['Sales'].apply(lambda x:0.85*x if x>200 else x)
df_sample.head(10)
```

# Concat vs Merge vs Join
- Concat -> take rows from one, adding rows together into one longer dataframe, add rows together in the same structure
- Merge -> represent database join, specify column
- Join -> based on index, joining into wider table on a shared key
- Merge and join are only for merging on a shared column
- Appending rows on each other -> concat
- Merge -> function of the pandas module
- Join -> method of a dataframe object
